```python
A = np.array([
    [6.26, 1.10, 0.98, 1.25],
    [1.10, 4.16, 1.00, 0.16],
    [0.98, 1.00, 5.44, 2.12],
    [1.25, 0.16, 2.12, 6.00]
])
```

## Mетод Данилевського

```python
def danilevsky(A, decimals=4):
    
    A = A.astype(float)
    n = A.shape[0]

    print("Початкова матриця A:\n", A, "\n")

    for k in range(n-1, 0, -1):
        
        print(f"КРОК {n-k}")
        
        M = np.eye(n)
        M_inv = np.eye(n)

        for j in range(n):
            if j == k-1:
                M[k-1, j] = 1 / A[k, k-1]
                M_inv[k-1, j] = A[k, k-1]
            else:
                M[k-1, j] = -A[k, j] / A[k, k-1]
                M_inv[k-1, j] = A[k, j]

        print("M:")
        print(M)
        print("\nM^{-1}:")
        print(M_inv)

        A = M_inv @ A @ M

        print("\nНова матриця A:")
        print(A)
        print("-"*50)

    A[np.abs(A) < 1e-10] = 0
    A_rounded = np.round(A, decimals)

    print("\nМатриця Фробеніуса:")
    print(A_rounded)

    return A_rounded
```

## Коефіцієнти характеристичного полінома

```python
def characteristic_polynomial(F):
    
    p = F[0]
    print("\nКоефіцієнти характеристичного полінома:")
    print(p)

    print("\nХарактеристичне рівняння:")
    print(f"x^4 - ({p[0]:.4f})x^3 - ({p[1]:.4f})x^2 - ({p[2]:.4f})x - ({p[3]:.4f}) = 0")

    return lambda x: x**4 - p[0]*x**3 - p[1]*x**2 - p[2]*x - p[3]
```

## Метод хорд

```python
def chord_method(f, a, b, e):
    c_prev = a  
    iteration = 0

    print(f"\nІнтервал [{a}, {b}]")
    print("i | x_k              | |x_k - x_k-1|")
    print("------------------------------------------------")

    while True:
        c = (a*f(b) - b*f(a)) / (f(b) - f(a))
        iteration += 1
        
        print(f"{iteration:4d} | {c:16.10f} | {abs(c - c_prev):.10f}")
        
        if abs(c - c_prev) < e or abs(f(c)) < e:
            break
        
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
        
        c_prev = c
    
    return c
```

## Пошук всіх власних чисел

```python
def intervals(f, a, b, step=0.5):
    intervals = []
    x = a

    while x < b:
        if f(x)*f(x+step) < 0:
            intervals.append((x, x+step))
        x += step

    return intervals
```

```python
def roots(f):
    
    intervals_list = intervals(f, 0, 10, step=0.5)
    roots = []

    for a, b in intervals_list:
        if f(a)*f(b) < 0:
            root = chord_method(f, a, b, 1e-6)
            roots.append(root)

    return roots
```
