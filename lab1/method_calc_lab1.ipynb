## $$ x^5 - 3x^4 + 7x^2 - 3 = 0 $$

```python
def f(x):
    return x**5 - 3*x**4 + 7*x**2 - 3

def df(x):
    return 5*x**4 - 12*x**3 + 14*x
```

## Метод бiсекцiї: 
### $$ c = \frac{a + b}{2} $$
- f - функція
- a, b - кінці інтервалу, який містить лише один корінь функції
- e - точність

```python
def bisection_method(f, a, b, e):

    iteration = 0
    
    print("Метод бісекції ")
    print("i | x_k              | (b - a)/2")
    print("------------------------------------------")
    
    while (b - a) / 2 > e:
        
        c = (a + b) / 2
        iteration += 1
        print(f"{iteration:4d} | {c:16.10f} | {c:.10f}")
        
        if f(a) * f(c) <= 0:
            b = c
        else:
            a = c

    root = (a + b) / 2
    
    return root, iteration
```

## Метод хорд
### $$c = \frac{a f(b) - b f(a)}{f(b) - f(a)} $$
- f - функція
- a, b - кінці інтервалу, який містить лише один корінь функції
- e - точність

```python
def chord_method(f, a, b, e):

    c_prev = a  
    iteration = 0
    print("Метод хорд ")
    print("i | x_k              | |x_k - x_k-1|")
    print("------------------------------------------------")
    
    while True:

        c = (a*f(b) - b*f(a)) / (f(b) - f(a))
        iteration += 1
        print(f"{iteration:4d} | {c:16.10f} | {abs(c - c_prev):.10f}")
        
        if abs(c - c_prev) < e or abs(f(c)) < e:
            break
        
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
        
        c_prev = c
    
    return c, iteration
```

## Метод Ньютона (дотичних):
### $$ x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)} $$
- f - функція 
- df - поххідна функції
- x0 - початкове наближення
- e - точність 

```python
def newton_method(f, df, x0, e):
    
    x_k = x0
    iteration = 0
    print("\nМетод Ньютона")
    print("i | x_k              | |x_k - x_k-1|")
    print("------------------------------------------------")
    
    while True:

        x_k_next = x_k - f(x_k) / df(x_k)
        iteration += 1
        
        print(f"{iteration:4d} | {x_k_next:16.10f} | {abs(x_k_next - x_k):.10f}")
        
        if abs(x_k_next - x_k) < e or abs(f(x_k_next)) < e:
            return x_k_next, iteration
        
        x_k = x_k_next
```

### $$ x^5 - 3x^4 + 7x^2 - 3 = 0 $$
### $$ [0; 1],  [-0,9; 0],  [-1; 0] $$
### - послідовність проміжків, кожен із яких містить лише один дійсний корінь рівняння

```python
intervals = [
    (-1.5, -1),
    (-0.9, -0.5),
    (0.5 , 1)
]
```

## Результати:

```python
e = 1e-5

for i, (a, b) in enumerate(intervals):

    print(f"\nІнтервал {i + 1 }: [{a}, {b}]")

    root_1, iterations_1 = bisection_method(f, a, b, e)
    root_2, iterations_2 = chord_method(f, a, b, e)
    root_3, iterations_3 = newton_method(f, df, a + 0.01 , e)
    
    print(f"\nМетод бісекції:      root = {root_1:.6f}, iterations = {iterations_1}")
    print(f"Метод хорд:          root = {root_2:.6f}, iterations = {iterations_2}")
    print(f"Метод Ньютона:       root = {root_3:.6f}, iterations = {iterations_3}")
```
