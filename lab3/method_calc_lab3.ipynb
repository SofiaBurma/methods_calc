### Система лінійних рівнянь

\begin{cases}

0.66x_1 + 0.44x_2 + 0.22x_3 + 1.00x_4 = 0.9
\end{cases}

## Матриця приведенна до діагональної переваги

```python
A = np.array([
    [0.822, -0.162, 0.084, 0.36],
    [-0.08,  0.79,  0.05,  0.11],
    [0.04,   0.11,  0.73, -0.11],
    [0.66,   0.44,  0.22,  1.0]
])

b = np.array([0.3, 0.5, 0.7, 0.9])
```

## Метод простої ітерації
### $$ \mathbf{x} = C \mathbf{x} + \mathbf{d} $$
### $$ с_{ij} = -\frac{a_{ij}}{a_{ii}}, \quad i \ne j $$
### $$ x^{(k+1)} = Cx^{(k)} + d $$
### $$ r = \lvert b - Ax \rvert $$

```python
def simple_iteration(A, b, e):

    n = len(b)

    C = np.zeros((n, n))
    d = np.zeros(n)

    for i in range(n):
        for j in range(n):
            if i != j:
                C[i][j] = -A[i][j] / A[i][i]
        d[i] = b[i] / A[i][i]

    x = np.zeros(n)

    print("k | x | критерій | нев'язка")

    for k in range(100):
        
        x_new = np.dot(C, x) + d

        criterion = np.max(np.abs(x_new - x))
        r = np.abs(b - np.dot(A, x_new))

        print( f"{k+1:2d} | "f"[{' '.join(f'{xi:.4f}' for xi in x_new)}] | "f"{criterion:.4f} | "f"[{' '.join(f'{ri:.4f}' for ri in r)}]" )
        
        if criterion < e:
            break

        x = x_new

    return x_new
```

## Результат

```python
x = simple_iteration(A, b, 1e-4)

print("\nРозв'язок системи:")
print(x)
```
